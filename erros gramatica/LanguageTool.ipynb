```python
errosNaoDetec = []

def contem_erro(texto):
    data = {
        "text": texto,
        "language": "pt-BR",
        "disabledRules": "UPPERCASE_SENTENCE_START,WHITESPACE_RULE,COMMA_PARENTHESIS_WHITESPACE"
    }
    response = requests.post(url, headers=headers, data=data)
    matches = response.json().get("matches", [])
    matches = [
        m for m in matches 
        if m['rule']['id'] not in [
            "COMMA_PARENTHESIS_WHITESPACE",
            "UPPERCASE_SENTENCE_START",
            "WHITESPACE_RULE"
        ]
    ]
    return len(matches) > 0

for index, row in df.iterrows():
    errosNaoDetec.append(not contem_erro(row['Frase_com_Erro']))

df['erro_nao_detectado'] = errosNaoDetec
```
